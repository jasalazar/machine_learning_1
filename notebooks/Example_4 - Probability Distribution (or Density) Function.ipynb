### Uniform Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
b = 100
n = 2000
data = np.random.uniform( a, b, n )
```

```python
%matplotlib inline
plt.hist( data )
```

## Normal Distribution

```python
data = np.random.randn( 10000 )
```

```python
x = range( 1, 10001 )
plt.plot( x, data )
```

```python
%matplotlib inline
plt.hist( data )
```

```python
plt.plot( x, sorted(data) ) #This proves it is a Gausian dist. Data is sorted.
```

```python
mu = 5.5
sd = 2.5
z = np.random.randn( 10000 )
data = mu + sd*z #X = mu + sd * z
plt.hist( data )
```
