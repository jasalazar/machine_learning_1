# Creating a data subset

```python
account_length = data[ "Account Length" ] #By selecting one column only, an object of type series is created
account_length.head()
```

```python
subset = data[[ "Account Length", "Phone", "Eve Charge", "Day Calls" ]] #Creating a subset out of the dataset. By selecting many columns, a dataframe is created
subset.head()
```

```python
desired_cols = [ "Account Length", "Phone", "Eve Charge", "Night Calls" ]
subset = data[ desired_cols ]
subset.head()
```

```python
all_cols_list = data.columns.values.tolist()
all_cols_list
```

```python
sublist = [ x for x in all_cols_list if x not in desired_cols ] #Sublist of non-desired (or target) columns
sublist
```

```python
subset = data[ sublist ] #Creating another subset with the desired cols
subset.head( 10 )
```

```python
subset[:8] #Subset of the first 8 rows
```
